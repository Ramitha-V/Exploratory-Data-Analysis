### Bar Plot

```python
tips
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with a size of 12x6 inches
plt.figure(figsize=(12, 6))

# Subplot 1: Seaborn barplot
plt.subplot(1, 2, 1)  
sns.barplot(x='sex', y='tip', data=tips)
plt.title("Sex vs Tips (Seaborn)")  

# Subplot 2: Matplotlib bar plot
plt.subplot(1, 2, 2)  
plt.bar(tips['sex'], tips['tip'])
plt.xlabel('Sex')
plt.ylabel('Tips')
plt.title('Sex vs Tips (Matplotlib)')  

plt.tight_layout() 
plt.show()
```

### Countplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with a size of 12x6 inches
plt.figure(figsize=(12, 6))

# Subplot 1: Seaborn countplot
plt.subplot(1, 2, 1)  
sns.countplot(x='sex', data=tips)
plt.title("Sex vs Count (Seaborn)")  

# Subplot 2: Matplotlib countplot
#Matplotlib does not have a countplot function of its own so we plot the values counts
sex_counts = tips['sex'].value_counts()

plt.subplot(1, 2, 2)
plt.bar(sex_counts.index, sex_counts.values)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Sex vs Count (Matplotlib)')

plt.tight_layout()
plt.show()
```

### BoxPlot

```python
#seaborn
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
#matplotlib
Days = ['Thur', 'Fri', 'Sat', 'Sun']
order_by_day = [tips[tips['day'] == day]['total_bill'].values for day in Days]

plt.figure(figsize=(8, 6))
plt.boxplot(order_by_day, labels=Days)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Day vs Total Bill (Matplotlib)')
plt.show()
```

```python
#Seaborn Horizontal 
sns.boxplot(data=tips,orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

```python
# Matplotlib
plt.figure(figsize=(8, 6))
plt.boxplot(tips['total_bill'], vert=False)
plt.xlabel('Total Bill')
plt.title('Boxplot of Total Bill (Matplotlib)')
plt.show()
```

### Violin Plot

```python
#Seaborn
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
plt.title('Violin Plot of Total Bill Amount by Day')
```

```python
#Matplotlib
# Group the data by 'day' and extract 'total_bill' values for each group
days = tips['day'].unique()
data_by_day = [tips[tips['day'] == day]['total_bill'].values for day in days]
plt.violinplot(data_by_day, showmedians=True)

plt.xticks(range(1, len(days) + 1), days)

# Set labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')
plt.title('Violin Plot of Total Bill Amount by Day')

# Show the plot
plt.show()
```

```python
#Seaborn
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
#Matplotlib
# Group the data by 'day' and 'sex' and extract 'total_bill' values for each group
groups = tips.groupby(['day', 'sex'])
data_by_group = [group['total_bill'].values for _, group in groups]

plt.figure(figsize=(10, 6))
plt.violinplot(data_by_group, showmedians=True)

# Set x-axis labels
days = tips['day'].unique()
day_labels = [f'{day} ({sex})' for day in days for sex in tips['sex'].unique()]
plt.xticks(range(1, len(day_labels) + 1), day_labels, rotation=45)

# Set labels and title
plt.xlabel('Day of the Week (Sex)')
plt.ylabel('Total Bill Amount')
plt.title('Violin Plot of Total Bill Amount by Day and Sex')

# Show the plot
plt.tight_layout()
plt.show()
```
