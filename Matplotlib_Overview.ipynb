```python
import pandas as pd
import numpy as np
# Importing matplotlib library
import matplotlib.pyplot as plt
```

Matplotlib


*   %matplotlib inline: plots are generated together with the code and not as separate file


```python
x=np.linspace(0,5,11)
```

```python
x
```

```python
y=x**2
```

Basic Matplotlib commands

```python
plt.plot(x,y,'r') #color red
plt.xlabel('x Axis title here')
plt.ylabel('y Axis title here')
plt.title("String Title Here")
plt.show()
```

Object Oriented Method

Creating Multiplot on same canvas

```python
#plt.subplot(nrows,ncols,plot_numeber)
plt.subplot(1,2,1)
plt.plot(x,y,'r--') #color r and type of line --
plt.xlabel("x")
plt.ylabel("y")
plt.title("x vs y")
plt.subplot(1,2,2)
plt.plot(y,x,'g*-') #color g and type of line *-
plt.xlabel("y")
plt.ylabel("x")
plt.title("y vs x")
plt.show()
```

```python
# Creating Figure(empty canvas)
fig=plt.figure()

# Add set of axes tp figure
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height (range 0 to 1)

# Plot on that set axes
axes.plot(x,y,'b')
axes.set_xlabel("Set X label")
axes.set_ylabel("Set Y label")
axes.set_title("Set Title")
```

```python
# Creating a blank canvas
fig1=plt.figure()

axes1=fig1.add_axes([0.1,0.1,0.8,0.8]) #main axes
axes2=fig1.add_axes([0.2,0.5,0.4,0.3]) #inset axes

#Larger Figure axes-1
axes1.plot(x,y,'b')
axes1.set_xlabel("Set X label_1")
axes1.set_ylabel("Set Y label_1")
axes1.set_title("Set Title_1")

#Insert Figure axes-2
axes2.plot(y,x,'r')
axes2.set_xlabel("Set X label_2")
axes2.set_ylabel("Set Y label_2")
axes2.set_title("Set Title_2");
```

subplots()

```python
#Tuple Packing to grab fig and axes
fig,axes=plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('x vs y');
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
# Axes is an array of axes to plot on
axes
```

```python
# We can iterate through this array

for ax in axes:
  ax.plot(x,y,'b')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title("Title")
# Display the figure object
fig
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

for ax in axes:
  ax.plot(x,y,'b')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title("Title")
# Display the figure object
fig
plt.tight_layout() #representing each graphs are an individual unit
```

**Figure size, aspect ratio and DPI**


*   figure size: width and height of the figure
*   dpi: dots-per-inch(pixel per inch)



```python
fig=plt.figure(figsize=(8,4),dpi=100)

plt.plot(x,y,'b')
```

```python
fig,axes=plt.subplots(figsize=(12,3))

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title("title")
```

Legends, Labels and Titles

```python
#Legends

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="x**2")
ax.set_xlabel("x")
ax.plot(x,x**3,label="x**3")
ax.set_ylabel("x**2/x**3")
ax.set_title("Title")
ax.legend()
```

Colors with MatLab like syntax

```python
fig,ax=plt.subplots()
ax.plot(x,x**2,'b--',label="x**2")
ax.plot(x,x**3,'r.-',label="x**3")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Title")
ax.legend()
```

Colors with color parameter

```python
fig,ax=plt.subplots()

ax.plot(x,x+1,color="blue",alpha=0.5) #half transparent
ax.plot(x,x+2,color="#8B008B") #RGB hex code
ax.plot(x,x+3,color="#FF8C00") #RGB hex code
```

### Line and marker styles

```python
fig,ax=plt.subplots(figsize=(12,6))
#line width options
ax.plot(x,x+1,color="red",linewidth=0.25)
ax.plot(x,x+2,color="red",linewidth=0.50)
ax.plot(x,x+3,color="red",linewidth=1.00)
ax.plot(x,x+4,color="red",linewidth=2.00)

#linestyle options
ax.plot(x,x+5,color="green",lw=3,linestyle='-')
ax.plot(x,x+6,color="green",lw=3,ls="-.")
ax.plot(x,x+7,color="green",lw=3,ls=":")

#Custom dash
line,=ax.plot(x,x+8,color="black",lw=1.50)
line.set_dashes([5,10,15,10])

# Possible marker symbols
ax.plot(x,x+9,color="blue",lw=3,ls='-',marker='+')
ax.plot(x,x+10,color="blue",lw=3,ls='--',marker='o')
ax.plot(x,x+11,color="blue",lw=3,ls='-',marker='s')
ax.plot(x,x+12,color="blue",lw=3,ls='--',marker="1")

#marker size and color
ax.plot(x,x+13,color="purple",lw=1,ls='-',marker='o',markersize=2)
ax.plot(x,x+14,color="purple",lw=1,ls='-',marker='o',markersize=4)
ax.plot(x,x+15,color="purple",lw=1,ls='-',marker='o',markersize=8,markerfacecolor="red")
ax.plot(x,x+16,color="purple",lw=1,ls='-',marker='s',markersize=8,markerfacecolor="yellow",markeredgewidth=3,markeredgecolor="green")

plt.show()
```
