```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

#### Correlation with Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

#### JointPlot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

#### Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

#### Distplot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

#### Categorical Plots

```python
# Set the style of seaborn
sns.set_style("whitegrid")

# Plot
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='sex', data=df, palette='Set2')

# Customize labels and title
ax.set_xlabel("Sex", fontsize=14)
ax.set_ylabel("Count", fontsize=14)
ax.set_title("Distribution of Sex", fontsize=16)

# Show plot
plt.show()
```

```python
# Set the style of seaborn
sns.set_style("whitegrid")

# Plot
plt.figure(figsize=(8, 6))
ax = sns.countplot(y='sex', data=df, palette='Set2')

# Customize labels and title
ax.set_xlabel("Sex", fontsize=14)
ax.set_ylabel("Count", fontsize=14)
ax.set_title("Distribution of Sex", fontsize=16)

# Show plot
plt.show()
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
# Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='smoker', y='total_bill', data=df)

# Customize labels and title
plt.xlabel("Smoker", fontsize=14)
plt.ylabel("Total Bill", fontsize=14)
plt.title("Boxplot of Total Bill by Smoker", fontsize=16)

# Show plot
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
