```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import math
```

```python
x=np.arange(0,math.pi*2,0.05)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y);
plt.grid()
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
```

```python
#Barplot: ax.bar(x,height,width,bottom,align)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=['c','c++','Java','Python','PHP']
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.grid()
plt.show()
```

```python
#Multiple bar plots
data=[[30,25,50,20],
     [40,23,51,17],
     [35,22,45,19]]
x=np.arange(4)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(x+0.00,data[0],color='b',width=0.25)
ax.bar(x+0.25,data[1],color='g',width=0.25)
ax.bar(x+0.50,data[2],color='r',width=0.25)
plt.grid()
plt.show()
```

```python
# Stacked Bar
N=5
menMeanscore=(20,35,20,25,27)
womenMeanscore=(25,32,34,20,25)
ind=np.arange(N)
width=0.35
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(ind,menMeanscore,width,color='r')
ax.bar(ind,womenMeanscore,width,bottom=menMeanscore,color='b')
ax.set_ylabel("Scores")
ax.set_title("Scores by group and gender")
ax.set_xticks(ind,('G1','G2','G3','G4','G5'))
ax.set_yticks(np.arange(0,81,10))
ax.legend(labels=['Men','Women'])
plt.show()
```

```python
# Histogram
fig,ax=plt.subplots(1,1)
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a,bins=[0,25,58,75,100])
ax.set_title("Histogram of Result")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel("Marks")
ax.set_ylabel("No of Students")
plt.show()
```
