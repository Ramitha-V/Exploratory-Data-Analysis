# Grid Plot

```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#grid() function
# Plotting subplots
x=np.arange(1,11)
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.plot(x,x**3,'r',lw=2)
plt.grid()
plt.title("default Grid")

plt.subplot(1,3,2)
plt.plot(x,np.exp(x),'b')
plt.grid(color='g',ls="-",lw="1")
plt.title("Custom Grid")

plt.subplot(1,3,3)
plt.plot(x,x)
plt.title("No Grid")

plt.tight_layout()
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.spines['bottom'].set_color('b')
ax.spines['left'].set_color('g')
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)

ax.plot([1,2,3,10,5])
plt.show()
```

### Customised Pair Plot

```python
from matplotlib import rcParams


sns.set_palette("cubehelix")
g=sns.PairGrid(iris,hue='species')
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
g.fig.suptitle("Customized PairGrid Plot for Iris Dataset",color="Gray",fontsize="20")

# Set custom font settings
font = {'family': 'serif', 'size': 15}
rcParams['font.family'] = font['family']
rcParams['font.size'] = font['size']


# List of custom labels for each subplot

x_labels = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]
y_labels = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]



# Set custom labels for each subplot
for i in range(len(g.axes)):
    g.axes[i, 0].set_ylabel(y_labels[i],color="gray")
    g.axes[-1, i].set_xlabel(x_labels[i],color="gray")

g.fig.tight_layout()
g.add_legend()

plt.show()
```

### Tips dataset

```python
tips = sns.load_dataset('tips')
tips.head()
```

### FacetGrid

```python
# Total bills are plotted as hist using map 
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
plt.tight_layout()
```

```python
# Scatter plot of tips using map 
g = sns.FacetGrid(tips, col="time",  row="smoker", hue='sex')
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

### Adding Grid Lines

```python
#Dummy Data
x1=np.linspace(0.0,5.0)
y1=np.cos(2*np.pi*x1)*np.exp(-1)

plt.figure(figsize=(12,6))
#Without grid
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.title("Plot Without Grid")

#With grid
plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.grid()
plt.title("Plot with Grid")


plt.tight_layout()
plt.show()
```

```python
# Line properties for Grid
x=np.linspace(0,2*np.pi,400)
y=np.sin(x**2)

plt.plot(x,y,'r')
plt.title("Plot with linewidth and linstyle")
plt.grid(color="grey",linewidth="1",linestyle="-")

plt.show()
```
