```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
#Create a figure object called fig using plt.figure() 
#Use add_axes to add an axis to the figure canvas at [0,0,1,1]. 
#Call this new axis ax. Plot (x,y) on that axes and set the labels and title.
fig=plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x=np.array(df['total_bill'])
y=np.array(df['tip'])
ax.plot(x, y,'go')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Title-total_bill & tip')
```

```python
#Create a figure object called fig using plt.figure() 
#Use add_axes to add an axis to the figure canvas at [0,0,1,1]. 
#Call this new axis ax. Plot (x,y) on that axes and set the labels and title.
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.hist(x, bins=10, color='skyblue')
ax2.hist(y, bins=10, color='salmon')

ax1.set_xlabel('Total Bill')
ax1.set_ylabel('Frequency')
ax1.set_title('Total Bill Histogram')

ax2.set_xlabel('Tip Amount')
ax2.set_ylabel('Frequency')
ax2.set_title('Tip Amount Histogram')

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Sex_encoded'] = label_encoder.fit_transform(df['sex'])
```

```python
#Create the plot by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]
#Use plt.subplots(nrows=1, ncols=2) to create the plot.
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

x=df['Sex_encoded']
y=df['tip']

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax1.bar(x,y, color='blue')
ax1.set_xlabel('Sex')
ax1.set_ylabel('Tips')
ax1.set_title('Bar Plot on ax1')

ax2.hist(y, bins=5, color='red')
ax2.set_xlabel('bins')
ax2.set_ylabel('Frequency')
ax2.set_title('Histogram on ax2')

plt.tight_layout()
plt.show()
```
