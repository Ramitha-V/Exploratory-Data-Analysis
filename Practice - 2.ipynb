```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# Distribution Plot
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.distplot(iris["sepal_length"], kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Sepal Length")
plt.grid()


sns.distplot(iris["sepal_width"], kde=True, ax=axes[0, 1])
axes[0, 1].set_title("Sepal Width")
plt.grid()

sns.distplot(iris["petal_length"], kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Petal Length")
plt.grid()

sns.distplot(iris["petal_width"], kde=True, ax=axes[1, 1])
axes[1, 1].set_title("Petal Width")
plt.grid()

plt.tight_layout()

plt.show()
```

```python
# Pair Plot
sns.set(style="ticks")
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])

plt.show()
```

```python
# Grid Plot- Pair Grid
g = sns.PairGrid(iris, hue="species", palette="rainbow")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
```

```python
# grid plot
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width")

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Scatterplot of Sepal Length vs. Sepal Width")

plt.grid(True)
plt.show()
```

```python
# Matrix Plot

plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(), annot=True)
plt.title("Matrix Plot")
plt.show()
```
